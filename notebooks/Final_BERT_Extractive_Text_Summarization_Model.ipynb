```
#RUN THIS ONE
from tqdm import tqdm_notebook
from summarizer import Summarizer

model=Summarizer()

extractive_summaries_list=[]
for i in tqdm_notebook(range(0,1849)):
  result=model(CNN_FINAL_TEST['article'].iloc[i], min_length=25,ratio=0.2)
  extractive_summaries_list.append(result)
```

Calcuate ROUGE Scores for BERT Model

```
from rouge_score import rouge_scorer

def calc_rouge(scorer, sources, predictions):
  rouge1 = 0
  rouge2 = 0
  rougeL = 0
  for i in range(len(sources)):
    scores = scorer.score(sources[i], predictions[i])
    rouge1+=scores['rouge1'][2]
    rouge2+=scores['rouge2'][2]
    rougeL+=scores['rougeL'][2]
  return rouge1/len(sources), rouge2/len(sources), rougeL/len(sources)
```

Final calculation

```
#RUN
#calculate ROUGE scores
ROUGE_score_list=[]
for i in range(0,len(extractive_summaries_list)):
  small_list=[]
  scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2','rougeL'], use_stemmer=True)
  sources = [extractive_summaries_list[i]]
  predictions = [CNN_FINAL_TEST['highlights'].iloc[i]]
  rouge1,rouge2,rougeL=calc_rouge(scorer, sources, predictions)
  small_list.append(rouge1)
  small_list.append(rouge2)
  small_list.append(rougeL)
  ROUGE_score_list.append(small_list)
```

Each list inside big list goes: rouge1, rouge2, rougeL

```
#print small_list
print(ROUGE_score_list)
```

```
#RUN calculate ROUGE scores for BERT. IGNORE OUTPUT FOR NOW
import numpy as np
rouge_1_list=[]
rouge_2_list=[]
rouge_L_list=[]
for item in ROUGE_score_list:
  rouge_1_list.append(item[0])
  rouge_2_list.append(item[1])
  rouge_L_list.append(item[2])

#average all the scores 
Bert_mean_rouge_1=np.mean(rouge_1_list)
Bert_mean_rouge_2=np.mean(rouge_2_list)
Bert_mean_rouge_L=np.mean(rouge_L_list)
print('BERT MEAN ROUGE 1', Bert_mean_rouge_1)
print('BERT MEAN ROUGE 2', Bert_mean_rouge_2)
print('BERT MEAN ROUGE L', Bert_mean_rouge_L)
  
```
